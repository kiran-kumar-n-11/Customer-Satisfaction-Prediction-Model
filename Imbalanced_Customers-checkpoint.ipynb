```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import RobustScaler
```

```python
def preprocess(train_df,test_df):
    
    time_cols = ['allot_time','order_time','pickup_time','delivered_time']
    for i in time_cols:
        train_df[i] = pd.to_datetime(train_df[i])
        test_df[i] = pd.to_datetime(test_df[i])
    print("converted to date time")
    for df in [train_df,test_df]:
        df['ref_allot_time'] = (df['allot_time'] - df['order_time']).dt.total_seconds()
        df['ref_pickup_time'] = (df['pickup_time'] - df['allot_time']).dt.total_seconds()
        df['ref_deliver_time'] = (df['delivered_time'] - df['pickup_time']).dt.total_seconds()
        df.drop(time_cols,axis=1,inplace=True)
        df.drop(['order_id','transport_id'],axis=1,inplace=True)
    print("created new cols")
    cat_cols = ['gender','membership','product_importan']
    label_encoder = LabelEncoder()
    for col in cat_cols:
        train_df[col]=label_encoder.fit_transform(train_df[col])
        test_df[col]=label_encoder.transform(test_df[col])
    print("label encoded")
    scale_cols = ['first_mile_distance','last_mile_distance','lifetime_order_count','customer_care_calls']
    robust_scaler = RobustScaler()
    for col in scale_cols:
        train_df[col] = robust_scaler.fit_transform(train_df[[col]])
        test_df[col] = robust_scaler.transform(test_df[[col]])
    print("Scaled!")
    
    return (train_df,test_df)
```

```python
train_df,test_df = preprocess(train_df,test_df)
```

```python
def convert_cat(df,df2):
    
    median=df['ref_allot_time'].median()
    q1=df['ref_allot_time'].quantile(0.25)
    q3=df['ref_allot_time'].quantile(0.75)
    iqr=(q3-q1)
#     mask = df['ref_allot_time']>median+1.5*iqr
    mask = df['ref_allot_time'].isna()
    df.loc[mask, 'cat_allot_time'] = 1
    df.loc[~mask, 'cat_allot_time'] = 0
#     mask = df2['ref_allot_time']>median+1.5*iqr
    mask = df2['ref_allot_time'].isna()
    df2.loc[mask, 'cat_allot_time'] = 1
    df2.loc[~mask, 'cat_allot_time'] = 0
    
    median=df['ref_deliver_time'].median()
    q1=df['ref_deliver_time'].quantile(0.25)
    q3=df['ref_deliver_time'].quantile(0.75)
    iqr=(q3-q1)
#     mask = (df['ref_deliver_time']>median+1.5*iqr) | (df['ref_deliver_time']<median-1.5*iqr)
    mask = df['ref_deliver_time'].isna()
    df.loc[mask, 'cat_deliver_time'] = 1
    df.loc[~mask, 'cat_deliver_time'] = 0
#     mask = (df2['ref_deliver_time']>median+1.5*iqr) | (df2['ref_deliver_time']<median-1.5*iqr)
    mask = df2['ref_deliver_time'].isna()
    df2.loc[mask, 'cat_deliver_time'] = 1
    df2.loc[~mask, 'cat_deliver_time'] = 0
    
    median=df['ref_pickup_time'].median()
    q1=df['ref_pickup_time'].quantile(0.25)
    q3=df['ref_pickup_time'].quantile(0.75)
    iqr=(q3-q1)
    mask = (df['ref_pickup_time']>median+1.5*iqr) | (df['ref_pickup_time']<median-1.5*iqr)
    mask = df['ref_pickup_time'].isna()
    df.loc[mask, 'cat_pickup_time'] = 1
    df.loc[~mask, 'cat_pickup_time'] = 0
#     mask = (df2['ref_pickup_time']>median+1.5*iqr) | (df2['ref_pickup_time']<median-1.5*iqr)
    mask = df2['ref_pickup_time'].isna()
    df2.loc[mask, 'cat_pickup_time'] = 1
    df2.loc[~mask, 'cat_pickup_time'] = 0
    
    df['frac_delivered'] = df['delivered_orders']/df['alloted_orders']
    df2['frac_delivered'] = df2['delivered_orders']/df2['alloted_orders']
    df.drop(['ref_pickup_time','ref_deliver_time','ref_allot_time','delivered_orders','alloted_orders',
             'undelivered_orders'],axis=1,inplace=True)
    df2.drop(['ref_pickup_time','ref_deliver_time','ref_allot_time','delivered_orders','alloted_orders',
             'undelivered_orders'],axis=1,inplace=True)
    df.fillna(df.mean(),inplace=True)
    df2.fillna(df2.mean(),inplace=True)
```
